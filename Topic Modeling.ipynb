```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer # Vectorizer - Text to numbers
from sklearn.decomposition import LatentDirichletAllocation # Topic Model

# Visualize Topic Models
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
```

```python
cfpb_df = pd.read_csv('cfpb/Consumer_Complaints.csv', usecols = [0, 1, 5], low_memory = False)
```

```python
cfpb_df.dropna(inplace=True)
```

```python
cfpb_df.shape
```

```python
cfpb_df.head()
```

```python
cfpb_df.Product.value_counts()
```

```python
student_loans_df = cfpb_df[cfpb_df.Product == 'Student loan']
```

```python
student_loans_df.columns = ['Date_received', 'Product', 'Consumer_complaint_narrative']
```

```python
vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 0.5, 
                                min_df = 10)
tfidf_matrix = vectorizer.fit_transform(student_loans_df.Consumer_complaint_narrative)
```

```python
tfidf_matrix.shape
```

```python
lda_tfidf = LatentDirichletAllocation(n_components=20, random_state=0)
lda_tfidf.fit(tfidf_matrix)
```

```python
pyLDAvis.sklearn.prepare(lda_tfidf, tfidf_matrix, vectorizer)
```
